```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
school_data_complete.head()
```

```python
# school_data_complete.head()
percent_math = 0
percent_read = 0
percent_both = 0
school_names = school_data['school_name']
district_sum = pd.DataFrame(columns = ['Total Schools', 'Total Students', 'Total Budget',
                                                'Average Math', 'Average Reading', 'Percent Pass Math',
                                                'Percent Pass Reading', 'Percent Pass Overall'])
total_schools = len(school_names)
total_students = school_data['size'].sum()
total_budget = school_data['budget'].sum()
# avg_math = sum(school_data_complete["math_score"])/len(school_data_complete["math_score"])
# avg_read = sum(school_data_complete["reading_score"])/len(school_data_complete["reading_score"])
avg_math = school_data_complete["math_score"].mean()
avg_read = school_data_complete["reading_score"].mean()
for index, row in school_data_complete.iterrows():
    if (row["math_score"] > 70):
        percent_math = percent_math + 1
    if (row["reading_score"] > 70):
        percent_read = percent_read + 1
    if (row["math_score"] > 70) & (row["reading_score"] > 70):
        percent_both = percent_both + 1
district_sum.loc[0] = [total_schools, total_students, total_budget, avg_math, avg_read, percent_math*100/total_students,
                percent_read*100/total_students, percent_both*100/total_students]
district_sum
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
school_sum_df = pd.DataFrame(columns = ['School Name','School Type', 'Total Students', 'Total School Budget', 
                                        'Per Student Budget','Average Math', 'Average Reading', 'Percent Pass Math',
                                        'Percent Pass Reading', 'Percent Pass Overall'])
#sort the dataframe by the values in the school name column
#school_df = school_data_complete.sort_values('school_name')
#reset the index so that the index is now based on the sorting
#school_df = school.df.reset_index(drop=True)
for i in range(total_schools):
    school_temp = school_data_complete.loc[school_data_complete['school_name'] == school_names[i], :]
    sch_percent_math = 0
    sch_percent_read = 0
    sch_percent_both = 0
    #sch_type = school_temp.loc[1,"type"]
    sch_students = school_data.loc[i,'size']
    sch_budget = school_data.loc[i, 'budget']
    avg_sch_math = school_temp["math_score"].mean()
    avg_sch_read = school_temp["reading_score"].mean()
    for index, row in school_temp.iterrows():
        if (row["math_score"] > 70):
            sch_percent_math = sch_percent_math + 1
        if (row["reading_score"] > 70):
            sch_percent_read = sch_percent_read + 1
        if (row["math_score"] > 70) & (row["reading_score"] > 70):
            sch_percent_both = sch_percent_both + 1
    school_sum_df.loc[i] = [school_names[i], school_temp.iloc[0,8], sch_students, sch_budget, sch_budget/sch_students, 
                            avg_sch_math, avg_sch_read, sch_percent_math*100/sch_students, sch_percent_read*100/sch_students, 
                            sch_percent_both*100/sch_students]
school_sum_new = school_sum_df.set_index('School Name')
school_sum_new
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#  School_data_complete has a column 'grade', "reading_score", "math_score", School Id, "school_name"
#create a pandas series for each grade
ninth = school_data_complete.loc[school_data_complete['grade'] == "9th", :]
tenth = school_data_complete.loc[school_data_complete['grade'] == "10th", :]
eleventh = school_data_complete.loc[school_data_complete['grade'] == "11th", :]
twelvth = school_data_complete.loc[school_data_complete['grade'] == "12th", :]
#ninth
#group each series by school
grouped_ninth_df = ninth.groupby(['school_name'])
grouped_tenth_df = tenth.groupby(['school_name'])
grouped_eleventh_df = eleventh.groupby(['school_name'])
grouped_twelvth_df = twelvth.groupby(['school_name'])
#combine the series into a dataframe
math_sum_df = pd.DataFrame(columns = ['9th','10th', '11th', '12th'])
math_sum_df.loc[:,'9th'] = grouped_ninth_df['math_score'].sum()/grouped_ninth_df['math_score'].count()
math_sum_df.loc[:,'10th'] = grouped_tenth_df['math_score'].sum()/grouped_tenth_df['math_score'].count()
math_sum_df.loc[:,'11th'] = grouped_eleventh_df['math_score'].sum()/grouped_eleventh_df['math_score'].count()
math_sum_df.loc[:,'12th'] = grouped_twelvth_df['math_score'].sum()/grouped_twelvth_df['math_score'].count()
math_sum_df
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
#combine the series into a dataframe
read_sum_df = pd.DataFrame(columns = ['9th','10th', '11th', '12th'])
read_sum_df.loc[:,'9th'] = grouped_ninth_df['reading_score'].sum()/grouped_ninth_df['reading_score'].count()
read_sum_df.loc[:,'10th'] = grouped_tenth_df['reading_score'].sum()/grouped_tenth_df['reading_score'].count()
read_sum_df.loc[:,'11th'] = grouped_eleventh_df['reading_score'].sum()/grouped_eleventh_df['reading_score'].count()
read_sum_df.loc[:,'12th'] = grouped_twelvth_df['reading_score'].sum()/grouped_twelvth_df['reading_score'].count()
read_sum_df
```
